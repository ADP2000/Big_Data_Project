## Precision, Recall, F-Measure for simple queries valutation

```python
true_positive_simple = 32
true_negative_simple = 8
false_positive_simple = 3
false_negative_simple = 6
```

```python
precision_simple = true_positive_simple / (true_positive_simple + false_positive_simple)
recall_simple = true_positive_simple / (true_positive_simple + false_negative_simple)
f_measure_simple = 2 * ((precision_simple * recall_simple) / (precision_simple + recall_simple))
print("Precision: \t" + str(precision_simple))
print("Recall: \t" + str(recall_simple))
print("F-Measure: \t" + str(f_measure_simple))
```

## Precision, Recall, F-Measure for intermediate queries valutation

```python
true_positive_intermediate = 21
true_negative_intermediate = 1
false_positive_intermediate = 3
false_negative_intermediate = 5
```

```python
precision_intermediate = true_positive_intermediate / (true_positive_intermediate + false_positive_intermediate)
recall_intermediate = true_positive_intermediate / (true_positive_intermediate + false_negative_intermediate)
f_measure_intermediate = 2 * ((precision_intermediate * recall_intermediate) / (precision_intermediate + recall_intermediate))
print("Precision: \t" + str(precision_intermediate))
print("Recall: \t" + str(recall_intermediate))
print("F-Measure: \t" + str(f_measure_intermediate))
```

## Precision, Recall, F-Measure for complex queries valutation

```python
true_positive_complex = 12
true_negative_complex = 2
false_positive_complex = 3
false_negative_complex = 3
```

```python
precision_complex = true_positive_complex / (true_positive_complex + false_positive_complex)
recall_complex = true_positive_complex / (true_positive_complex + false_negative_complex)
f_measure_complex = 2 * ((precision_complex * recall_complex) / (precision_complex + recall_complex))
print("Precision: \t" + str(precision_complex))
print("Recall: \t" + str(recall_complex))
print("F-Measure: \t" + str(f_measure_complex))
```

## Precision, Recall, F-Measure for total queries valutation

```python
total_true_positive = true_positive_simple + true_positive_intermediate + true_positive_complex
total_true_negative = true_negative_simple + true_negative_intermediate + true_negative_complex
total_false_positive = false_positive_simple + false_positive_intermediate + false_positive_complex
total_false_negative = false_negative_simple + false_negative_intermediate + false_negative_complex
```

```python
total_precision = total_true_positive / (total_true_positive + total_false_positive)
total_recall = total_true_positive / (total_true_positive + total_false_negative)
total_f_measure = 2 * ((total_precision * total_recall) / (total_precision + total_recall))
print("Precision: \t" + str(total_precision))
print("Recall: \t" + str(total_recall))
print("F-Measure: \t" + str(total_f_measure))
```

## Graphs for Precision, Recall and F-Measure

```python
import pandas as pd
```

```python
data = {
    "Type Valutation" : ["Simple", "Intermediate", "Complex", "Total"],
    "Precision": [precision_simple, precision_intermediate, precision_complex, total_precision],
    "Recall": [recall_simple, recall_intermediate, recall_complex, total_recall],
    "F-Measure": [f_measure_simple, f_measure_intermediate, f_measure_complex, total_f_measure]    
}

```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# Larghezza delle barre
bar_width = 0.20

# Posizioni delle barre
r1 = np.arange(len(df))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Crea il grafico
plt.figure(figsize=(10, 6))
plt.bar(r1, df['Precision'], color='blue', width=bar_width, edgecolor='grey', label='Precision')
plt.bar(r2, df['Recall'], color='red', width=bar_width, edgecolor='grey', label='Recall')
plt.bar(r3, df['F-Measure'], color='green', width=bar_width, edgecolor='grey', label='F-Measure')

# Aggiungi etichette e titolo
plt.xlabel('Type of Queries', fontweight='bold')
plt.ylabel('Scores', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(df))], df['Type Valutation'])
plt.title('Scores by Type of Queries', fontweight='bold')
plt.legend()
```
